```python
# Importar libreria pandas
import pandas as pd
# Leer CSV planta y guardar como DataFrame (agregar encoding='latin1', sep=';')
df = pd.read_csv("planta032025.csv", encoding='latin1', sep=';')
df.head()
```

```python
# Revisar columnas en forma de índice y tipo de datos
for i, (col, tipo) in enumerate(df.dtypes.items(), start=1):
    print(f"{i}. {col}: {tipo}")
```

```python
# REDUCCIÓN DE COLUMNAS, REEMPLAZO DE NAN EN ANTIGÜEDAD, LIMPIEZA DE DATOS NULOS, ELIMINACION DE TACA=0 Y CAT=600, Y CONVERSION A NUMERO DE "TACA", "TASA", "TotApPat", "Costo Laboral","TotDesc", "Neto", "A01", "A14"

# Lista de columnas a conservar
columnas_a_conservar = [
    "Año", "Mes", "Nombre", "CUIL", "Fecnacim", "Sexo", "Ministerio", "Centro",
    "Descri Centro", "Sector", "Descri Sector", "Padron", "Escalafón","Nombre_Escalafon",
    "Poder", "Agrup_Tramo", "Categ", "Antig", "Regimen", "TACA", "TASA",
    "TotApPat", "Costo Laboral", "TotDesc", "Neto", "A01", "A14"
]

# --- Paso 1: Reducir columnas (solo si existen)
existentes = [c for c in columnas_a_conservar if c in df.columns]
faltantes  = [c for c in columnas_a_conservar if c not in df.columns]
if faltantes:
    print("⚠️ No se redujo porque faltan columnas:", faltantes)
else:
    df = df[existentes].copy()

# --- Paso 2:
  # CUIL como string para no perder ceros a la izquierda ni formato
    if "CUIL" in df.columns:
        df["CUIL"] = df["CUIL"].astype(str).str.strip()

  # Antig: NaN -> -1 (solo esa columna)
  # Las celdas vacias representan un -1 para la liquidación, pero no se colocan.
    if "Antig" in df.columns:
        # si viene como texto, normalizo
        df["Antig"] = (df["Antig"].astype(str).str.replace(",", ".", regex=False))
        df["Antig"] = pd.to_numeric(df["Antig"], errors="coerce").fillna(-1)

  # Categ a número para poder comparar con 600
  # Se eliminan vacaciones no gozadas.
    if "Categ" in df.columns:
        df["Categ"] = pd.to_numeric(df["Categ"], errors="coerce")

  # Lista de columnas que deben ser numéricas
    columnas_numericas = [
    "TACA", "TASA", "TotApPat", "Costo Laboral", "TotDesc", "Neto", "A01", "A14"
    ]
  # Normalizar numéricos con coma decimal -> punto y convertir
    for col in columnas_numericas:
        if col in df.columns:
            df[col] = (
                df[col].astype(str)
                      .str.strip()
                      .str.replace(",", ".", regex=False)
            )
            df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0)

# --- Paso 3: Filtrar
  # Eliminar TACA <= 0
  # Se eliminan los puestos con cargos retenidos por Licencias sin goce y Licencia por cargo de mayor jerarquia.
    if "TACA" in df.columns:
        df = df[df["TACA"] > 0]

  # Eliminar Categ = 600
  # Se eliminan vacaciones no gozadas.
    if "Categ" in df.columns:
        df = df[df["Categ"] != 600]

# --- Paso 4: Eliminar Enfermeros
# Aseguramos que las columnas relevantes estén en numérico
df["Escalafón"] = pd.to_numeric(df["Escalafón"], errors="coerce")
df["Agrup_Tramo"] = pd.to_numeric(df["Agrup_Tramo"], errors="coerce")
df["A14"] = pd.to_numeric(
    df["A14"].astype(str).str.replace(",", ".", regex=False),
    errors="coerce"
).fillna(0)

# Eliminar enfermeros según las condiciones
# Escalafón=8: Ley 1148-Q
# Agrup_Tramo=302: Agrupamiento Tecnicos y Administrativos
# A14>0: Adicional de Peligrosidad, solo para personal Tecnico de la Salud
condicion = (df["Escalafón"] == 8) & (df["Agrup_Tramo"] == 302) & (df["A14"] >0)
df = df[~condicion]  # usamos ~ para excluir

# --- Paso 5: Guardar
df.to_csv("planta032025_reducido.csv", index=False, encoding="latin1", sep=";")
print("✅ Archivo reducido guardado como planta032025_reducido.csv")
```

```python
# ELIMINACION DE CUIL DUPLICADOS CON SUMA PREVIA DE TACA
import pandas as pd

# Leer CSV reducido
df = pd.read_csv("planta032025_reducido.csv", encoding="latin1", sep=";")

# --- Paso 1: asegurar que TACA sea numérico ---
df["TACA"] = pd.to_numeric(df["TACA"], errors="coerce").fillna(0)

# --- Paso 2: sumar TACA por CUIL ---
# Para que al calcular el TACA total por agente se tengan en cuenta los adicionales por comision de ss. y los jornales de campaña. Se calcula cuánto gana realmente cada agente.
taca_por_cuil = (
    df.groupby("CUIL", as_index=False)["TACA"]
      .sum()
      .rename(columns={"TACA": "TACA_total"})
)

# --- Paso 3: elegir la fila "sobreviviente" por CUIL, evitando Categ=580 ---
# Para poder evidenciar correctamente la Repartición/Ministerio de pertenencia del agente, es decir el CUIL queda asociado a la repartición correcta y no a la categoría 580.
df["_prior"] = (df["Categ"] == 580).astype(int)   # 0 si ≠580, 1 si =580

sobrevivientes = (
    df.sort_values(["CUIL", "_prior"])   # primero los no-580
      .drop_duplicates(subset="CUIL", keep="first")
      .drop(columns=["_prior"])
)

# --- Paso 4: fusionar con el TACA total ---
final = sobrevivientes.merge(taca_por_cuil, on="CUIL", how="left")

# Reemplazar la columna TACA por el total
final["TACA"] = final["TACA_total"]
final = final.drop(columns=["TACA_total"])

# --- Paso 5: guardar CSV final ---
final.to_csv("planta032025_sin_duplicados.csv", index=False, encoding="latin1", sep=";") #acá se guarda nuevo csv: planta032025_sin_duplicados.csv
print("✅ Archivo generado: planta032025_sin_duplicados.csv")
print(final[["CUIL", "TACA"]].head())  # en 'final' la TACA ya es el total

```

```python
# CÁLCULO DE EDAD A 2025 Y CLASIFICACIÓN POR RANGO ETARIO

import pandas as pd

# --- Leer el archivo limpio ---
df = pd.read_csv("planta032025_sin_duplicados.csv", encoding="latin1", sep=";")

# --- Calcular la edad en 2025 ---
df["anio_nac"] = df["Fecnacim"].astype(str).str[:4].astype(int)
df["edad_2025"] = 2025 - df["anio_nac"]

# --- Clasificar en rangos etários ---
# Rangos etarios: 18 a 49, 50 a 54, 55 a 59, 60 a 64, 65 a 69 y 70 o mas
def clasificar_edad(edad):
    if 18 <= edad <= 49:
        return "18 a 49"
    elif 50 <= edad <= 54:
        return "50 a 54"
    elif 55 <= edad <= 59:
        return "55 a 59"
    elif 60 <= edad <= 64:
        return "60 a 64"
    elif 65 <= edad <= 69:
        return "65 a 69"
    elif edad >= 70:
        return "70 o más"
    else:
        return "Menor de 18"

df["rango_etario"] = df["edad_2025"].apply(clasificar_edad)

# --- Clasificar antigüedad en rangos ---
# Rangos antiguedad: 30 O MAS , 25 A 29, 20 A 24, 15 A 19, 10 A 14, 5 A 9 y -1 A 4
def clasificar_antig(antig):
    if antig >= 30:
        return "30 O MAS"
    elif 25 <= antig <= 29:
        return "25 A 29"
    elif 20 <= antig <= 24:
        return "20 A 24"
    elif 15 <= antig <= 19:
        return "15 A 19"
    elif 10 <= antig <= 14:
        return "10 A 14"
    elif 5 <= antig <= 9:
        return "5 A 9"
    elif -1 <= antig <= 4:
        return "-1 A 4"
    else:
        return "Fuera de rango"

df["rango_antig"] = df["Antig"].apply(clasificar_antig)

# --- Vista previa ---
print(df[["Fecnacim", "edad_2025", "rango_etario", "Antig", "rango_antig"]].head(10))

# --- Guardar con las nuevas columnas ---
df.to_csv("planta032025_sin_duplicados_con_rangos_edadyantig.csv", index=False, encoding="latin1", sep=";")
```

```python
# ARMADO DE PROGRAMAS DE RETIRO

import pandas as pd

# =========================
# Configuración de entrada, definiendo que archivo se va a leer y como se va a guardar el resultado
# =========================
IN_FILE  = "planta032025_sin_duplicados_con_rangos_edadyantig.csv"
OUT_FILE = "planta032025_clasificado_retiro_voluntario.csv"

# =========================
# Cargar y normalizar
# =========================
df = pd.read_csv(IN_FILE, encoding="latin1", sep=";")

# Normalizo campos clave
# Para que las columnas sean texto, sin espacios de más
for c in ["rango_antig", "rango_etario", "Sexo"]:
    if c in df.columns:
        df[c] = df[c].astype(str).str.strip()

# Sexo a formato M/F
# En esta base ahora no hace falta, pero normaliza si aparece HOMBRE o MUJER, sea M y F
df["Sexo_norm"] = (
    df["Sexo"].str.upper()
      .replace({"HOMBRE": "M", "MUJER": "F"})
      .fillna(df["Sexo"].str.upper())
)

# Aseguro columnas de salida (para guardar los resultados)
for col in ["CONDICIÓN", "PROPUESTA", "APERTURA", "CONDICIÓN_FINAL"]:
    if col not in df.columns:
        df[col] = None

# Helper para trabajar en mayúsculas. Es decir, se crean atajos
ANTIG = df["rango_antig"].str.upper()
EDAD  = df["rango_etario"].str.upper()
SEXO  = df["Sexo_norm"].str.upper()

# =========================
# 1) JUB EDAD AVANZADA
# =========================
m0 = df["CONDICIÓN"].isna()
mask_jub_av = (
    m0 & EDAD.eq("70 O MAS") & ANTIG.isin(["10 A 14", "15 A 19", "20 A 24", "25 A 29"])
)
df.loc[mask_jub_av, ["CONDICIÓN","PROPUESTA","APERTURA","CONDICIÓN_FINAL"]] = \
    ["JUB EDAD AVANZADA", "-", "JUB EDAD AVANZADA", "-"]

# =========================
# 2) JUBILABLES
# =========================
m0 = df["CONDICIÓN"].isna()
mask_jubilables = m0 & ANTIG.eq("30 O MAS") & EDAD.isin(["65 A 69", "70 O MAS"])
df.loc[mask_jubilables, ["CONDICIÓN","PROPUESTA","APERTURA","CONDICIÓN_FINAL"]] = \
    ["JUBILABLES", "-", "JUBILABLES", "-"]

# =========================
# 3) RETIRO PROGRAMADO
# =========================
m0 = df["CONDICIÓN"].isna()

mask_rp_70 = m0 & ANTIG.isin(["10 A 14", "15 A 19", "20 A 24"]) & EDAD.eq("65 A 69")
df.loc[mask_rp_70, ["CONDICIÓN","PROPUESTA","APERTURA","CONDICIÓN_FINAL"]] = \
    ["RETIRO PROGRAMADO", "70% NETO", "70% NETO H/ 70 AÑOS", "HASTA CUMPLIR 70"]

mask_rp_75_70 = m0 & ANTIG.eq("25 A 29") & EDAD.eq("65 A 69")
df.loc[mask_rp_75_70, ["CONDICIÓN","PROPUESTA","APERTURA","CONDICIÓN_FINAL"]] = \
    ["RETIRO PROGRAMADO", "75% NETO", "75% NETO / JOR", "HASTA CUMPLIR 70"]

mask_rp_75_65 = m0 & ANTIG.eq("25 A 29") & EDAD.eq("60 A 64")
df.loc[mask_rp_75_65, ["CONDICIÓN","PROPUESTA","APERTURA","CONDICIÓN_FINAL"]] = \
    ["RETIRO PROGRAMADO", "75% NETO", "75% NETO / JOR", "HASTA CUMPLIR 65"]

mask_rp_75_60F = m0 & ANTIG.eq("25 A 29") & EDAD.eq("55 A 59") & SEXO.eq("F")
df.loc[mask_rp_75_60F, ["CONDICIÓN","PROPUESTA","APERTURA","CONDICIÓN_FINAL"]] = \
    ["RETIRO PROGRAMADO", "75% NETO", "75% NETO H/ 60 AÑOS", "HASTA CUMPLIR 60"]

mask_rp_80_65 = m0 & ANTIG.eq("30 O MAS") & EDAD.eq("60 A 64")
df.loc[mask_rp_80_65, ["CONDICIÓN","PROPUESTA","APERTURA","CONDICIÓN_FINAL"]] = \
    ["RETIRO PROGRAMADO", "80% NETO", "80% NETO H/ JOR", "HASTA CUMPLIR 65"]

mask_rp_80_60F = m0 & ANTIG.eq("30 O MAS") & EDAD.eq("55 A 59") & SEXO.eq("F")
df.loc[mask_rp_80_60F, ["CONDICIÓN","PROPUESTA","APERTURA","CONDICIÓN_FINAL"]] = \
    ["RETIRO PROGRAMADO", "80% NETO", "80% NETO H/ 60 AÑOS", "HASTA CUMPLIR 60"]

# =========================
# 4) RETIRO ANTICIPADO
# =========================
m0 = df["CONDICIÓN"].isna()

sueldos_anticipado = {
    "-1 A 4":  "18 SUELDOS",
    "5 A 9":   "23 SUELDOS",
    "10 A 14": "28 SUELDOS",
    "15 A 19": "30 SUELDOS",
    "20 A 24": "33 SUELDOS",
    "25 A 29": "35 SUELDOS",
    "30 O MAS":"38 SUELDOS",
}

mask_ra = m0 & ANTIG.isin(list(sueldos_anticipado.keys()))
df.loc[mask_ra, "CONDICIÓN"] = "RETIRO ANTICIPADO"
df.loc[mask_ra, "PROPUESTA"] = ANTIG.map(sueldos_anticipado)
df.loc[mask_ra, "APERTURA"]  = df.loc[mask_ra, "PROPUESTA"]
df.loc[mask_ra, "CONDICIÓN_FINAL"] = "-"

# =========================
# 5) Calcular años de acompañamiento
# =========================
df["años_acompañamiento"] = None

df.loc[df["CONDICIÓN_FINAL"].eq("HASTA CUMPLIR 70"), "años_acompañamiento"] = 70 - df["edad_2025"]
df.loc[df["CONDICIÓN_FINAL"].eq("HASTA CUMPLIR 65"), "años_acompañamiento"] = 65 - df["edad_2025"]
df.loc[df["CONDICIÓN_FINAL"].eq("HASTA CUMPLIR 60"), "años_acompañamiento"] = 60 - df["edad_2025"]

# =========================
# Guardar
# =========================
out_cols = ["rango_antig","rango_etario","Sexo","edad_2025",
            "CONDICIÓN","PROPUESTA","APERTURA","CONDICIÓN_FINAL","años_acompañamiento"]

df.to_csv(OUT_FILE, index=False, encoding="latin1", sep=";")
print("✅ Archivo generado:", OUT_FILE)
print(df[out_cols].head(12))
```
